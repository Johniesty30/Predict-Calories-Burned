## Converting categorical features

```python
le2 = LabelEncoder()
df["Workout_Type"]  = le2.fit_transform(df["Workout_Type"])
print("Workout_type mapping", dict(enumerate(le2.classes_)))
df["Gender"]  = le2.fit_transform(df['Gender'])
print("Gender mapping", dict(enumerate(le2.classes_)))
```

## Splitting data (target- inputs)

```python
y= df["Calories_Burned"]
x = df.drop(columns=["Calories_Burned"])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(x_train, y_train)
```

## Training

```python
models = {
    "RandomForestRegressor" :RandomForestRegressor(),
}
```

```python
for name,model in models.items():
    model.fit(x_train,y_train)
    score = model.score(x_test,y_test)
    print(f"{name} score : {score * 100:.2f}%")
```

## Cross-Validation

```python
crossval = cross_val_score(model, x,y,cv=4)
crossval
```
